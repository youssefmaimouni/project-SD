```python
import pandas as pd
df=pd.read_csv('skynews_full_data.csv')
df
```

supression de la colonne indice Unnamed: 0

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df.isnull().sum()
```

supresion des 4 ligne vide de la colonne description 

```python
df.dropna(subset=['description'], inplace=True)
```

```python
df.isnull().sum()
```

```python
df
```

supression des lien image

```python
    df.drop(columns=['image_url'], inplace=True)
```

```python
df
```

conversion en format DATE la colonne published at

```python
df['published_at'] = pd.to_datetime(df['published_at'], utc=True)
```

```python
df['published_at'] = df['published_at'].dt.tz_localize(None)
```

```python
df
```

Séparation des dates

```python
df['year'] = df['published_at'].dt.year
df['month'] = df['published_at'].dt.month
df['day'] = df['published_at'].dt.day
```

```python
df
```

```python
df.drop(columns=['published_at'], inplace=True)
```

```python
df
```

conversion des Tags Liste en texre 

```python
import ast
df['tags_text'] = df['tags'].apply(lambda tags: ' '.join(ast.literal_eval(tags)) if isinstance(tags, str) else '')
```

```python
df
```

```python
df.drop(columns=['tags'], inplace=True)
```

```python
df
```

```python
df['month'].unique()
```

```python
df['day'].unique()
```

```python
df['published_at'] = pd.to_datetime(df[['year', 'month', 'day']])
```

```python
df
```

```python
df.drop(columns=['year', 'month', 'day'], inplace=True)
```

```python
df
```

```python
df.rename(columns={'tags_text': 'tags', 'description': 'content'}, inplace=True)
```

```python
df
```

```python
df['source'] = df['source'].str.replace(' ', '_')
```
