```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup

options = Options()
options.add_argument("--headless")  # Run in the background
options.add_argument("--disable-blink-features=AutomationControlled")  # Avoid bot detection

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
driver.get("https://www.france24.com/ar/آسيا/")

html = driver.page_source
driver.quit()
# Parse the page source with BeautifulSoup
soup = BeautifulSoup(html, "html.parser")
soup
```

```python

# Extract all news headlines
headlines = soup.find_all("h2", class_="")

# Print extracted headlines
for headline in headlines:
    print(headline.text.strip())
```

```python
# Look for containers holding articles, such as divs or sections
containers = soup.find_all("div", class_="m-item-list-article__wrapper")  # Replace with actual class name
print(containers[0],"\n +++++++++++++++++++++++++++++++++++++++++++++++ \n")
# Check if it contains any articles
for container in containers:
    print(container.prettify())  # Print the HTML of each container
```

```python
articles = soup.find_all("div", class_="m-item-list-article__wrapper")
articles[0]
```

```python
tags = soup.find_all("div", class_="a-tag-section")
for tag in tags:
    print(tag.text)
```

```python
for article in articles:
    # Extract the link to the article
    link_tag = article.find("a", href=True)
    article_link = f"https://www.france24.com{link_tag['href']}" if link_tag else "No Link"
    
    # Extract the article title
    title_tag = article.find("div", class_="article__title")
    title = title_tag.text.strip() if title_tag else "No Title"
    
    # Extract the summary or "chapo" of the article
    chapo_tag = article.find("div", class_="article__chapo")
    summary = chapo_tag.text.strip() if chapo_tag else "No Summary"
    
    # Print the extracted information
    print(f"📰 Title: {title}")
    print(f"🔗 Link: {article_link}")
    print(f"✍️ Summary: {summary}\n")
```

```python
import requests
from bs4 import BeautifulSoup
import csv


# Open a CSV file to save the data
with open("france24_arabe_articles_extended_asia.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    # Write the header
    writer.writerow(["source", "author?", "title", "content", "description?", "url", "image_url?", "published_at", "tags"])

    for article in articles:
        # Source
        source = "France 24"  # Since it's from France24

        # Author (if available)
        author_tag = article.find("span", class_="article__author")  # Check for author class, modify if necessary
        author = author_tag.text.strip() if author_tag else "No Author"

        # Title
        title_tag = article.find("div", class_="article__title")
        title = title_tag.text.strip() if title_tag else "No Title"

        # Content (This may need deeper inspection, modify if necessary)
        content_tag = article.find("div", class_="article__content")  # Modify class as needed
        content = content_tag.text.strip() if content_tag else "No Content"

        # Description (Using summary/chapo)
        description_tag = article.find("div", class_="article__chapo")
        description = description_tag.text.strip() if description_tag else "No Description"

        # URL
        link_tag = article.find("a", href=True)
        article_link = f"https://www.france24.com{link_tag['href']}" if link_tag else "No URL"

        # Image URL (if available)
        image_tag = article.find("img", class_="m-figure__img")
        image_url = image_tag["src"] if image_tag else "No Image URL"

        # Published At (check if available, may need further inspection)
        published_at_tag = article.find("time", class_="article__published")  # Modify class if necessary
        published_at = published_at_tag["datetime"] if published_at_tag else "No Published Date"

        # Tags (check if available)
        tags_tag = article.find("div", class_="a-tag-section")  # Modify if necessary
        tags = tags_tag.text.strip() if tags_tag else "No Tags"

        # Write the extracted data to CSV
        writer.writerow([source, author, title, content, description, article_link, image_url, published_at, tags])

print("✅ Data saved to france24_articles_extended.csv")
```
