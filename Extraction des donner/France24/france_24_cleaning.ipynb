```python
import pandas as pd

# Load the CSV file
df = pd.read_csv("fichier_fusionne_2025_2024.csv")  # Replace with your actual file name


```

```python
df
```

```python
df = df[df['published_at'] != 'No Published Date']
```

```python
df.drop(columns=['Unnamed: 0', 'author', 'image_url?'], inplace=True)
```

```python
df.dropna(subset=['description'], inplace=True)
```

```python
df
```

```python
df.drop(columns='content',inplace=True)
```

```python
df.isnull().sum()
```

```python
df.tags.values
```

```python

# Nettoyage des tags : supprimer les espaces, sauts de ligne et caractères spéciaux inutiles
df["tags"] = df["tags"].astype(str).str.strip().str.replace(r"\s+", " ", regex=True)

# Compter les occurrences après nettoyage
tag_counts = df["tags"].value_counts().to_dict()

# Afficher le dictionnaire propre
print(tag_counts)
```

```python
df.tags.isnull().sum()
```

```python
df["content"] = df["description"].astype(str).str.strip().str.replace(r"\s+", " ", regex=True)
df.drop(columns="description", inplace=True)
```

```python
df
```

```python
df.columns
```

```python
from datetime import datetime
def normalize_date(date):

     return datetime.strptime(date, "%d/%m/%Y - %H:%M").strftime("%Y-%m-%d")

df["published_at"] = df["published_at"].apply(normalize_date)
```
