```python
import pandas as pd
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les stopwords arabes
nltk.download("stopwords")
from nltk.corpus import stopwords

arabic_stopwords = set(stopwords.words("arabic"))

# Charger les donnÃ©es
df = pd.read_csv("full_data_fusionne.csv")
df = df.dropna(subset=['title'])  # Supprimer les lignes vides

# PrÃ©traitement du texte
def preprocess_text(text):
    text = text.lower()  # Minuscule
    text = "".join(c for c in text if c.isalnum() or c.isspace())  # Supprimer la ponctuation
    words = text.split()  # Tokenizer
    words = [word for word in words if word not in arabic_stopwords]  # Supprimer stopwords
    return " ".join(words)

df["cleaned_title"] = df["title"].apply(preprocess_text)

# TF-IDF sur les articles
vectorizer = TfidfVectorizer(max_features=1000)
X = vectorizer.fit_transform(df["cleaned_title"])
```

```python
inertia = []
K_range = range(15, 35)

for k in K_range:
    model = KMeans(n_clusters=k, random_state=42, n_init=10)
    model.fit(X)
    inertia.append(model.inertia_)

# Tracer le graphique
plt.figure(figsize=(8, 5))
sns.lineplot(x=list(K_range), y=inertia, marker="o")
plt.xlabel("Nombre de clusters")
plt.ylabel("Inertie (Inertia)")
plt.title("MÃ©thode du coude pour choisir K")
plt.show()
```

```python
optimal_k =98  # Remplace par la valeur trouvÃ©e
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df["cluster"] = kmeans.fit_predict(X)

# Afficher les articles regroupÃ©s par cluster
for i in range(optimal_k):
    print(f"\nðŸ”¹ Cluster {i}:")
    print(df[df["cluster"] == i][["title", "content"]])  # Afficher 5 articles par cluster
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA

# Appliquer K-Means avec le nombre optimal de clusters trouvÃ©
labels = kmeans.fit_predict(X)  # X est ta matrice de caractÃ©ristiques

# âœ… 1. Score de Silhouette
silhouette_avg = silhouette_score(X, labels)
print(f"ðŸ”¹ Score de silhouette: {silhouette_avg:.3f}")

# âœ… 2. Inertie (Within-cluster sum of squares - WCSS)
inertia = kmeans.inertia_
print(f"ðŸ”¹ Inertie du modÃ¨le: {inertia:.2f}")

# âœ… 3. Visualisation des Clusters en 2D avec PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X.toarray())  # Si X est un sparse matrix, utilise toarray()

plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=labels, palette="tab10", s=50)
plt.xlabel("Composante principale 1")
plt.ylabel("Composante principale 2")
plt.title("Visualisation des Clusters (PCA)")
plt.legend(title="Cluster")
plt.show()
```

```python
# Test suggested k values
for k in [30, 28]:
    model = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = model.fit_predict(X)
    score = silhouette_score(X, labels)
    print(f"k={k}: Silhouette={score:.3f}, Inertia={model.inertia_:.1f}")
```

```python
from sklearn.metrics import silhouette_score

best_k = 0
best_score = -1
for k in range(99, 130):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    if score > best_score:
        best_score = score
        best_k = k

print(f"Meilleur k selon le score de silhouette : {best_k} -- {score}")
```
