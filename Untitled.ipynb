```python
import pandas as pd
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les stopwords arabes
nltk.download("stopwords")
from nltk.corpus import stopwords

arabic_stopwords = set(stopwords.words("arabic"))

# Charger les données
df = pd.read_csv("full_data_fusionne.csv")
df = df.dropna(subset=['content'])  # Supprimer les lignes vides

# Prétraitement du texte
def preprocess_text(text):
    text = text.lower()  # Minuscule
    text = "".join(c for c in text if c.isalnum() or c.isspace())  # Supprimer la ponctuation
    words = text.split()  # Tokenizer
    words = [word for word in words if word not in arabic_stopwords]  # Supprimer stopwords
    return " ".join(words)

df["cleaned_content"] = df["content"].apply(preprocess_text)

# TF-IDF sur les articles
vectorizer = TfidfVectorizer(max_features=1000)
X = vectorizer.fit_transform(df["cleaned_content"])
```

```python
inertia = []
K_range = range(2, 10)

for k in K_range:
    model = KMeans(n_clusters=k, random_state=42, n_init=10)
    model.fit(X)
    inertia.append(model.inertia_)

# Tracer le graphique
plt.figure(figsize=(8, 5))
sns.lineplot(x=list(K_range), y=inertia, marker="o")
plt.xlabel("Nombre de clusters")
plt.ylabel("Inertie (Inertia)")
plt.title("Méthode du coude pour choisir K")
plt.show()
```

```python
optimal_k = 4  # Remplace par la valeur trouvée
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df["cluster"] = kmeans.fit_predict(X)

# Afficher les articles regroupés par cluster
for i in range(optimal_k):
    print(f"\n🔹 Cluster {i}:")
    print(df[df["cluster"] == i][["title", "content"]])  # Afficher 5 articles par cluster
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA

# Appliquer K-Means avec le nombre optimal de clusters trouvé
kmeans = KMeans(n_clusters=7, random_state=42, n_init=10)
labels = kmeans.fit_predict(X)  # X est ta matrice de caractéristiques

# ✅ 1. Score de Silhouette
silhouette_avg = silhouette_score(X, labels)
print(f"🔹 Score de silhouette: {silhouette_avg:.3f}")

# ✅ 2. Inertie (Within-cluster sum of squares - WCSS)
inertia = kmeans.inertia_
print(f"🔹 Inertie du modèle: {inertia:.2f}")

# ✅ 3. Visualisation des Clusters en 2D avec PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X.toarray())  # Si X est un sparse matrix, utilise toarray()

plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=labels, palette="tab10", s=50)
plt.xlabel("Composante principale 1")
plt.ylabel("Composante principale 2")
plt.title("Visualisation des Clusters (PCA)")
plt.legend(title="Cluster")
plt.show()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords

# Télécharger les mots vides en arabe
nltk.download('stopwords')
arabic_stopwords = list(stopwords.words('arabic'))  # Convertir en liste

# Convertir les données en DataFrame pour une manipulation facile
df = pd.read_csv("full_data_fusionne.csv")

# Étape 1 : Vectorisation des contenus avec TF-IDF
vectorizer = TfidfVectorizer(stop_words=arabic_stopwords)  # Utiliser une liste de mots vides
X = vectorizer.fit_transform(df['content'])  # Utilisez 'content' pour le clustering

# Étape 2 : Clustering avec K-means
num_clusters = 3  # Nombre de clusters (à ajuster selon vos besoins)
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(X)

# Étape 3 : Ajouter les clusters aux données
df['Cluster'] = kmeans.labels_

# Afficher les résultats
print(df[['title', 'Cluster']])

# Étape 4 : Visualisation des clusters (optionnel)
# Réduction de dimension avec PCA pour visualiser en 2D
pca = PCA(n_components=2)
reduced_features = pca.fit_transform(X.toarray())

# Affichage des clusters en 2D
plt.figure(figsize=(10, 6))
for i in range(num_clusters):
    plt.scatter(reduced_features[df['Cluster'] == i, 0], reduced_features[df['Cluster'] == i, 1], label=f'Cluster {i}')

plt.title('تجميع المقالات باستخدام خوارزمية K-means')
plt.xlabel('المكون الرئيسي 1')
plt.ylabel('المكون الرئيسي 2')
plt.legend()
plt.show()
```
