```python
import pandas as pd 

chemin_fichiers=['./Extraction des donner/France24/france_24_clened_data.csv','./Extraction des donner/Sky News/skynews_data_cleaned_vf.csv','./Extraction des donner/Al Hura/alhura_full_data_cleaned.csv','./Extraction des donner/Al Jazera/aljazira_cleaned.csv']
dataframes = []

for chemin_fichier in chemin_fichiers :
    df = pd.read_csv(chemin_fichier)
    dataframes.append(df)

df_fusionne = pd.concat(dataframes, ignore_index=True)
df_fusionne
```

```python
df_fusionne.loc[df_fusionne['tags'] == 'NO_TAGS', 'tags'] = 'no_tags'
df_fusionne[df_fusionne['tags']=='no_tags']
```
