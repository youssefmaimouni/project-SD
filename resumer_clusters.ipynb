```python
# ⚠️ Exécute cette cellule en premier dans ton notebook
import os
os.environ["TRANSFORMERS_NO_TF"] = "1"

# Ensuite, importe les bibliothèques
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import pandas as pd

# Lire les données
df = pd.read_csv("random_1000_clusters.csv")

# Choisir un cluster
cluster_id = 0
texts = df[df['cluster'] == cluster_id]['full_text'].dropna().tolist()

# Concaténer les textes
full_text = " ".join(texts)
full_text = full_text[:2000]  # Limiter à 2000 caractères

# Charger le modèle mT5 multilingue
model_name = "csebuetnlp/mT5_multilingual_XLSum"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# Tokeniser et résumer
input_ids = tokenizer("summarize: " + full_text, return_tensors="pt", truncation=True, max_length=512).input_ids
output_ids = model.generate(input_ids, max_length=150, num_beams=4, early_stopping=True)

# Résultat
summary = tokenizer.decode(output_ids[0], skip_special_tokens=True)
print("📝 Résumé en arabe :")
print(summary)
```
