```python
import pandas as pd
import matplotlib.pyplot as plt
import arabic_reshaper
from bidi.algorithm import get_display

# Tes données (remarque : j'ai retiré la ligne avec 'NO tags')
data = {
    'Catégorie': [
        'أخبار إسرائيل', 'الأخبار المغاربية', 'الشرق الأوسط',
        'أخبار أميركا', 'أخبار فلسطين', 'أوروبا', 'أخبار لبنان', 'شرق أوسط',
        'إسرائيل', 'أمريكا', 'فلسطين', 'الرياضة', 'السياسة', 'الاقتصاد',
        'العلوم', 'الثقافة', 'التكنولوجيا', 'الأديان', 'الشخصيات'
    ],
    'Valeur': [
        4197, 3229, 2209, 1828, 
        1775, 1038, 924, 909, 
        887, 759, 680, 550, 412, 380,
        300, 278, 250, 150, 100
    ]
}

# Création du DataFrame
df = pd.DataFrame(data)

# Fonction de reshaping de texte arabe
def reshape_arabic(text):
    reshaped_text = arabic_reshaper.reshape(text)
    bidi_text = get_display(reshaped_text)
    return bidi_text

# Appliquer la fonction à chaque catégorie
df['Catégorie_affichage'] = df['Catégorie'].apply(reshape_arabic)

# Tri décroissant des valeurs
df_sorted = df.sort_values(by='Valeur', ascending=False)

# Tracé du graphique
plt.figure(figsize=(12, 8))
plt.barh(df_sorted['Catégorie_affichage'], df_sorted['Valeur'], color='skyblue')
plt.xlabel('Valeur', fontsize=12)
plt.ylabel('Catégories', fontsize=12)
plt.title('La Distribution Des Tags', fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.7)  # Correction de la syntaxe de la grille
plt.tight_layout()
plt.show()
```
