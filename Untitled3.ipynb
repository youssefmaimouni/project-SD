```python
import json
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from collections import Counter
from tqdm import tqdm
import nltk
from nltk.corpus import stopwords
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd

# Télécharger les mots d'arrêt en arabe
nltk.download('stopwords')
arabic_stop_words = list(stopwords.words('arabic'))  # Convertir en liste

# Étape 1 : Charger les données JSON
def load_data(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        data = json.load(file)
    return data

# Chemin vers votre fichier JSON
file_path = "random_sample_1000.csv"  # Remplacez par le chemin de votre fichier
data = pd.read_csv(file_path).to_dict('records')

# Extraire les titres, le contenu et les dates des articles
titles = [article['title'] for article in data]
contents = [article['content'] for article in data]
dates = [article['date'] for article in data]

# Étape 2 : Clustering avec K-means et TF-IDF
vectorizer = TfidfVectorizer(max_features=5000, stop_words=arabic_stop_words)
X = vectorizer.fit_transform(contents)

# Déterminer le nombre optimal de clusters (k) avec le score de silhouette
silhouette_scores = []
k_values = range(2, 20)  # Testez des valeurs de k de 2 à 20

for k in tqdm(k_values, desc="Calcul des scores de silhouette"):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    silhouette_avg = silhouette_score(X, labels)
    silhouette_scores.append(silhouette_avg)

# Trouver le k optimal (celui avec le score de silhouette le plus élevé)
optimal_k = k_values[silhouette_scores.index(max(silhouette_scores))]
print(f"Nombre optimal de clusters détecté : {optimal_k}")

# Tracer la courbe des scores de silhouette
plt.figure(figsize=(8, 5))
plt.plot(k_values, silhouette_scores, marker='o')
plt.axvline(x=optimal_k, color='r', linestyle='--', label=f'Optimal (k={optimal_k})')
plt.xlabel('Nombre de clusters (k)')
plt.ylabel('Score de silhouette')
plt.title('Score de silhouette pour choisir k')
plt.legend()
plt.show()

# Appliquer K-means avec le nombre optimal de clusters
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(X)

# Ajouter les labels des clusters aux données
for i, article in enumerate(data):
    article['cluster'] = int(kmeans.labels_[i])

# Étape 3 : Générer un résumé pour chaque cluster avec BERT
def generate_cluster_summary(contents, labels, n_words=10):
    # Charger un modèle BERT multilingue
    model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')
    
    # Encoder les contenus avec BERT
    content_embeddings = model.encode(contents)
    
    cluster_summaries = {}
    for cluster in range(optimal_k):
        # Extraire les indices des articles du cluster
        cluster_indices = [i for i, label in enumerate(labels) if label == cluster]
        
        # Calculer la moyenne des embeddings BERT pour le cluster
        cluster_embedding = np.mean(content_embeddings[cluster_indices], axis=0)
        
        # Trouver l'article le plus proche du centroïde du cluster
        similarities = cosine_similarity([cluster_embedding], content_embeddings[cluster_indices])
        most_representative_index = cluster_indices[np.argmax(similarities)]
        
        # Utiliser le contenu de l'article le plus représentatif comme résumé
        cluster_summaries[cluster] = contents[most_representative_index][:500]  # Limiter à 500 caractères
    
    return cluster_summaries

# Générer les résumés des clusters avec BERT
cluster_summaries = generate_cluster_summary(contents, kmeans.labels_)

# Afficher les résumés des clusters
for cluster, summary in cluster_summaries.items():
    print(f"Cluster {cluster} : {summary}")

# Étape 4 : Calculer le degré de similarité intra-cluster avec BERT
def calculate_intra_cluster_similarity(contents, labels):
    model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')
    content_embeddings = model.encode(contents)
    
    intra_cluster_similarities = {}
    for cluster in range(optimal_k):
        # Extraire les indices des articles du cluster
        cluster_indices = [i for i, label in enumerate(labels) if label == cluster]
        
        # Calculer la similarité cosinus moyenne entre les articles du cluster
        cluster_embeddings = content_embeddings[cluster_indices]
        similarities = cosine_similarity(cluster_embeddings)
        intra_cluster_similarities[cluster] = np.mean(similarities)
    
    return intra_cluster_similarities

# Calculer les similarités intra-cluster
intra_cluster_similarities = calculate_intra_cluster_similarity(contents, kmeans.labels_)

# Afficher les similarités intra-cluster
for cluster, similarity in intra_cluster_similarities.items():
    print(f"Degré de similarité intra-cluster {cluster} : {similarity:.4f}")

# Étape 5 : Stocker les résultats dans un fichier JSON
output_data = []
for article in data:
    output_data.append({
        'title': article['title'],
        'date': article['date'],
        'content': article['content'],
        'cluster': article['cluster']
    })

# Ajouter les résumés des clusters et les similarités intra-cluster au fichier JSON
output_data.append({
    'cluster_summaries': cluster_summaries,
    'intra_cluster_similarities': intra_cluster_similarities
})

with open('clustered_articles.json', 'w', encoding='utf-8') as file:
    json.dump(output_data, file, ensure_ascii=False, indent=4)

print("Les résultats ont été enregistrés dans 'clustered_articles.json'.")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import arabic_reshaper
from bidi.algorithm import get_display
import re

# Load the data
df = pd.read_csv('random_sample_1000.csv')

# Arabic text preprocessing
def preprocess_arabic(text):
    if not isinstance(text, str):
        return ""
    # Remove punctuation and special characters
    text = re.sub(r'[^\w\s]', ' ', text)
    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    return text

# Preprocess titles
df['cleaned_title'] = df['title'].apply(preprocess_arabic)

# Vectorization with CountVectorizer
vectorizer = CountVectorizer(max_features=1000)
X = vectorizer.fit_transform(df['cleaned_title'])

# Determine optimal number of clusters using silhouette score
silhouette_scores = []
for n_clusters in range(2, 11):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    silhouette_scores.append(silhouette_avg)
    print(f"For n_clusters = {n_clusters}, silhouette score is {silhouette_avg}")

# Plot silhouette scores
plt.plot(range(2, 11), silhouette_scores)
plt.title('Silhouette Score for Different Numbers of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.show()

# Apply K-Means with optimal number of clusters (choose based on silhouette score)
optimal_clusters = silhouette_scores.index(max(silhouette_scores)) + 2  # +2 because range starts at 2
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
df['cluster'] = kmeans.fit_predict(X)

# Visualize clusters using PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(X.toarray())

plt.figure(figsize=(12, 8))
for i in range(optimal_clusters):
    plt.scatter(principal_components[df['cluster'] == i, 0], 
                principal_components[df['cluster'] == i, 1], 
                label=f'Cluster {i}')

plt.title('K-Means Clustering of Arabic News Titles')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()

# Display sample titles from each cluster
for cluster_num in range(optimal_clusters):
    print(f"\nCluster {cluster_num} - Sample Titles:")
    cluster_titles = df[df['cluster'] == cluster_num]['title'].sample(min(5, len(df[df['cluster'] == cluster_num])))
    for title in cluster_titles:
        # Reshape and display Arabic text properly
        reshaped_text = arabic_reshaper.reshape(title)
        bidi_text = get_display(reshaped_text)
        print(f"- {bidi_text}")
```

```python
# Function to get BERT embeddings
def get_bert_embeddings(texts, batch_size=32):
    embeddings = []
    for i in tqdm(range(0, len(texts), batch_size)):
        batch = texts[i:i+batch_size]
        inputs = tokenizer(batch, return_tensors="pt", padding=True, truncation=True, max_length=128)
        with torch.no_grad():
            outputs = model(**inputs)
        # Use mean pooling to get sentence embeddings
        batch_embeddings = torch.mean(outputs.last_hidden_state, dim=1)
        embeddings.append(batch_embeddings)
    return torch.cat(embeddings, dim=0)

# Get embeddings for all titles
titles = df['title'].tolist()
embeddings = get_bert_embeddings(titles)

# Convert to numpy array
X = embeddings.numpy()
```

```python
# Determine optimal number of clusters using silhouette score
silhouette_scores = []
k_range = range(50, min(100, len(titles)))  # Testing 2-5 clusters (limited by sample size)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(X)
    if len(set(cluster_labels)) > 1:  # Silhouette score requires at least 2 clusters
        silhouette_avg = silhouette_score(X, cluster_labels)
        silhouette_scores.append(silhouette_avg)
    else:
        silhouette_scores.append(0)

# Plot silhouette scores
plt.figure(figsize=(10, 6))
plt.plot(k_range, silhouette_scores, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Different Cluster Numbers')
plt.show()

# Choose optimal k (highest silhouette score)
optimal_k = k_range[np.argmax(silhouette_scores)]
print(f"Optimal number of clusters: {optimal_k}")
print(np.argmax(silhouette_scores))
```
