```python
import pandas as pd 
```

```python
df=pd.read_csv("aljazira_articles.csv")
```

```python
df['source'] = 'aljazira'
df['no_[url]']=''
df['no_[tags]']=''
```

```python
df
```

```python
df.columns = ['title', 'content', 'published_at','source', 'no_[url]','no_[tags]']
```

```python
df = df[['source', 'title', 'no_[url]', 'published_at', 'no_[tags]','content']]
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True) 
```

```python
df.duplicated().sum()
```

```python
df.columns = df.columns.str.strip()  # Supprime les espaces au début/fin
```

```python
df
```

```python
df.to_csv('aljazira_cleaned.csv', index=False)
```
