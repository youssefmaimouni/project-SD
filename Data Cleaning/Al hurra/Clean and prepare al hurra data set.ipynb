```python
# Fill missing values according to your specifications
df['author'] = df['author'].fillna('Not mentioned')
df['image_url'] = df['image_url'].fillna('Not available')
```

```python
# For missing content, copy from description
df['content'] = df['content'].fillna(df['description'])
```

```python
df.isnull().sum()
```

```python
import pandas as pd
import re
from nltk.corpus import stopwords
import nltk
from collections import Counter

# Download Arabic stopwords
nltk.download('stopwords')
arabic_stopwords = set(stopwords.words('arabic'))

def extract_arabic_tags(row, num_tags=5, min_length=5):
    """Extract tags from Arabic text using title-content common words"""
    if isinstance(row['tags'], list) and len(row['tags']) > 0:
        return row['tags']
    
    # Combine title and content
    text = f"{row['title']} {row['content']}"
    
    # Arabic-specific cleaning
    text = re.sub(r'[^\u0600-\u06FF\u0750-\u077F\s]', '', text)  # Keep Arabic characters
    text = re.sub(r'\s+', ' ', text).strip()
    
    # Tokenization for Arabic
    tokens = re.findall(r'\b[\u0600-\u06FF]{3,}\b', text)  # Arabic words with 3+ letters
    
    # Filtering
    filtered_words = [
        word for word in tokens
        if word not in arabic_stopwords
        and len(word) >= min_length
    ]
    
    # Get most common words
    word_counts = Counter(filtered_words)
    return [word for word, count in word_counts.most_common(num_tags)]

# Apply to dataframe
df['tags'] = df.apply(extract_arabic_tags, axis=1)
```
