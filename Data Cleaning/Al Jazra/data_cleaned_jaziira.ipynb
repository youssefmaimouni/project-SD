```python
import pandas as pd 
```

```python
df=pd.read_csv("./../../Extraction des donner/Al Jazera/aljazira_articles.csv")
```

```python
df['source'] = 'aljazira'
df['url']='no_url'
df['tags']='no_tags'
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True) 
```

```python
df.duplicated().sum()
```

```python
df.columns = df.columns.str.strip()  # Supprime les espaces au début/fin
```

```python
df
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%Y/%m/%d').dt.strftime('%Y-%m-%d')
df
```

```python
df.rename(columns={'Date': 'published_at'}, inplace=True)
df.rename(columns={'Title': 'title'}, inplace=True)
df.rename(columns={'Content': 'content'}, inplace=True)
df
```

```python
df.to_csv('./../../Extraction des donner/Al Jazera/aljazira_cleaned.csv', index=False)
```
